```python
#Operators

#Addition:
print(3 + 2)

#Subtraction:
print(3 - 2)

#Multiplication:
print(3 * 2)

#Division:
print(3 / 2)

#Modulus:
print(3 % 2)

#Increment:
number = 10
number -= 5 #(note: 5 here is called a stepsize)
print(number)

#comparison operators: They return boolean output i.e. it only returns true or false.
print(3 > 5)
print(3 < 5)
#(x == 5) == means equal to while = means assigned to.
#(x != 5) != means not equal to
entry_age = 17
my_age = 17
result = my_age > entry_age
print(result)

entry_age = 17
my_age = 18
result = my_age > entry_age
print(result)

entry_age = 17
my_age = 17
result = my_age >= entry_age
print(result)

#Logical operators: or, and & xor
gender = 'Female'

comp_and = my_age >= entry_age and gender == 'Male'
print(comp_and) #True ==1, False == 0, and == 1 * 0

comp_and = my_age >= entry_age and gender == 'Female'
print(comp_and) #True ==1, False == 0, and == 1 * 0

comp_and = my_age >= entry_age or gender == 'Male'
print(comp_and) #True ==1, False == 0, or == 1 + 0

#Converting from one variable type to another
my_input = '6'
int(my_input) / 2

my_ans = 2 * 7
print('The answer is ' + str(my_ans))


#Assignment:

var_1 = '45'
var_2 = '2'
int(var_1)
int(var_2)
result = (int(var_1) + int(var_2))

print(f'{var_1} + {var_2} = '+ str(result))
```
